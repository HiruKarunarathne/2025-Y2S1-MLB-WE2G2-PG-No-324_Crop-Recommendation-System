```python
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

# Load Dataset
This section loads the dataset into a DataFrame.

```python
datasetPath = "/content/Crop_recommendation.csv"
```

```python
df = pd.read_csv("Crop_recommendation.csv")
```

# Before Handling Imbalance
Visualizing the class distribution before applying SMOTE.

```python
class_counts_before = df['label'].value_counts()

plt.figure(figsize=(12,6))
class_counts_before.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Class Distribution of Crops (Before Handling Imbalance)")
plt.xlabel("Crop Type")
plt.ylabel("Number of Samples")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
X = df.drop('label', axis=1) # Features:
y = df['label']              # Target
```

```python
class_counts_imbalanced = df['label'].value_counts() # Count samples per class
```

```python
smote = SMOTE(random_state=42)   # Import SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)
```

Create dataframe for resampled data

```python
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['label'] = y_resampled

class_counts_after = df_resampled['label'].value_counts()

plt.figure(figsize=(12,6))
class_counts_after.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Class Distribution of Crops (After Handling Imbalance with SMOTE)")
plt.xlabel("Crop Type")
plt.ylabel("Number of Samples")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

SAVE DATASETS

```python
df_resampled.to_csv("Crop_recommendation_balanced.csv", index=False)

print("Datasets saved as 'Crop_recommendation_imbalanced.csv' and 'Crop_recommendation_balanced.csv'")
```
