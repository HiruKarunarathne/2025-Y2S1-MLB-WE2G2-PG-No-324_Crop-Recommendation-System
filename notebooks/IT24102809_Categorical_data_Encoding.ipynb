```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load your dataset
df = pd.read_csv("Crop_recommendation_balanced.csv")

# 2. Check unique labels
print("Unique labels:", df['label'].unique())

# 3. Label encode the 'label' column
df['label_encoded'] = pd.factorize(df['label'])[0]  # assigns a unique number to each label

# 4. Show the first few rows with new labels
print(df[['label', 'label_encoded']].head())

# 5. Show the full mapping of label → number
label_mapping = dict(zip(df['label'].unique(), range(len(df['label'].unique()))))
print("Label mapping (label → number):")
for label, num in label_mapping.items():
    print(f"{label} → {num}")

# 6. EDA Visualization: Distribution of labels
df['label'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Labels")
plt.xlabel("Crop")
plt.ylabel("Number of Samples")
plt.show()

# 7. Save the updated dataset
df.to_csv("Crop_recommendation_label_encoded.csv", index=False)
```
