```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df=pd.read_csv("Crop_recommendation_label_encoded.csv")
```

```python
display(df.head())
```

```python
print(df.shape)
```

```python
plt.figure(figsize=(12, 6))  
df.boxplot()
plt.show()
```

```python
sns.boxplot(x=df['N'])
plt.show()

sns.boxplot(x=df['P'])
plt.show()

sns.boxplot(x=df['K'])
plt.show()

sns.boxplot(x=df['temperature'])
plt.show()

sns.boxplot(x=df['humidity'])
plt.show()

sns.boxplot(x=df['ph'])
plt.show()


sns.boxplot(x=df['rainfall'])
plt.show()
```

```python
#remove the outliers 1st
##identify numerical colum
numeric_cols=df.select_dtypes(include='number').columns

Q1=df[numeric_cols].quantile(0.25)
Q3=df[numeric_cols].quantile(0.75)
IQR=Q3-Q1

condition = ~((df[numeric_cols] <(Q1-1.5* IQR)) | (df[numeric_cols] > (Q3+ 1.5*IQR))).any(axis=1)


df_cleaned_1=df[condition]

print(df_cleaned_1.shape)
```

```python
plt.figure(figsize=(12, 6))  
df_cleaned_1.boxplot()
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['N'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['P'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['K'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['temperature'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['humidity'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['ph'])
plt.show()

plt.figure(figsize=(4, 3))
sns.boxplot(x=df_cleaned_1['rainfall'])
plt.show()
```

```python
print(df.shape)

print(df_cleaned_1.shape)
```

```python
#remove the outliers 2nd time
##identify numerical colum
numeric_cols=df_cleaned_1.select_dtypes(include='number').columns

Q1=df_cleaned_1[numeric_cols].quantile(0.25)
Q3=df_cleaned_1[numeric_cols].quantile(0.75)
IQR=Q3-Q1

condition = ~((df_cleaned_1[numeric_cols] <(Q1-1.5* IQR)) | (df_cleaned_1[numeric_cols] > (Q3+ 1.5*IQR))).any(axis=1)

#apply this condition full dataframe
df_cleaned_2=df_cleaned_1[condition]

print(df_cleaned_2.shape)
```

```python
plt.figure(figsize=(12, 6))  
df_cleaned_2.boxplot()
plt.show()
```

```python
#Remove the out lier 3rd time
##identify numerical colum
numeric_cols=df_cleaned_1.select_dtypes(include='number').columns

Q1=df_cleaned_2[numeric_cols].quantile(0.25)
Q3=df_cleaned_2[numeric_cols].quantile(0.75)
IQR=Q3-Q1

condition = ~((df_cleaned_2[numeric_cols] <(Q1-1.5* IQR)) | (df_cleaned_2[numeric_cols] > (Q3+ 1.5*IQR))).any(axis=1)

#apply this condition full dataframe
df_cleaned_3=df_cleaned_2[condition]

print(df_cleaned_3.shape)
```
